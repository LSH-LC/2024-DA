# 파일 데이터 처리
- 인구 데이터 전처리
- 인구 데이터 시각화

```python
# 파일을 읽고 pop에 저장

import pandas as pd

pop = pd.read_csv("202408popmonth.csv", encoding = 'cp949', index_col = 0, thousands = ',')
```

```python
# pop 읽기

pop
```

```python
# pop의 열이름 수정

pop.columns = ['총인구수','세대수','세대당인구','남자인구수','여자인구수','남여비율']
pop.head()
```

```python
# pop의 index 수정

pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
# 코랩 추천 그래프

pop
```

```python
# 총 인구수 산점도

import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90)
plt.show
```

```python
# 총 인구수 막대 그래프

import matplotlib.pyplot as plt
import seaborn as sns

pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수',data=pop1)
plt.xticks(rotation=90)
plt.show()
```
